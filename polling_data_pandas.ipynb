```python
data2016[['Address']] = data2016[['Address']].astype(str)
data2015[['Address']] = data2016[['Address']].astype(str)
data2014[['Address']] = data2016[['Address']].astype(str)

print("             2016  2015   2014")
print("boulevard:")
print("      ", data2016['Address'].str.contains('boulevard').any(),
                data2015['Address'].str.contains('boulevard').any(),
                data2014['Address'].str.contains('boulevard').any())

print("street:")
print("      ", data2016['Address'].str.contains('Street').any(),
                 data2015['Address'].str.contains('Street').any(),
                 data2014['Address'].str.contains('Street').any())  
   
print("Drive:")
print("      ", data2016['Address'].str.contains('Drive').any(),
                data2015['Address'].str.contains('Drive').any(),
                data2014['Address'].str.contains('Drive').any())
 
print("Avenue:")
print("      ", data2016['Address'].str.contains('Avenue').any(),
                data2015['Address'].str.contains('Avenue').any(),
                data2014['Address'].str.contains('Avenue').any())
  
print("Parkway:")
print("      ", data2016['Address'].str.contains('Parkway').any(),
                data2015['Address'].str.contains('Parkway').any(),
                data2014['Address'].str.contains('Parkway').any())

print("Lane:")
print("      ", data2016['Address'].str.contains('Lane').any(),
                data2015['Address'].str.contains('Lane').any(),
                data2014['Address'].str.contains('Lane').any())
  
```

## Transform feature: Precinct(s)
Separate entries with commas

If there is more than one there will be a comma,

if there is check if next caracter is a space

    if yes, do nothing
    
    if not, insert space

```python
print("Precinct(s) \n 2016",data2016['Precinct(s)'].str.contains(',').any())
print("               2015",data2016['Precinct(s)'].str.contains(',').any())
print("               2014",data2016['Precinct(s)'].str.contains(',').any())
```
